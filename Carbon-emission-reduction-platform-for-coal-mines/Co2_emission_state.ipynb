```python
years = np.array([2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023]).reshape(-1, 1)
```

```python
degree = 2
for state, emissions in data.items():
    emissions = np.array(emissions)


    poly = PolynomialFeatures(degree=degree)
    years_poly = poly.fit_transform(years)


    model = LinearRegression()
    model.fit(years_poly, emissions)


    future_years = np.array([2024, 2025, 2026]).reshape(-1, 1)
    future_years_poly = poly.transform(future_years)
    predicted_emissions = model.predict(future_years_poly)


    plt.figure(figsize=(10, 6))
    plt.scatter(years, emissions, color='blue', label='Actual Data')
    plt.plot(years, model.predict(years_poly), color='red', label='Polynomial Fit')
    plt.plot(future_years, predicted_emissions, color='green', linestyle='dashed', label='Predicted Future Emissions')
    plt.xlabel('Year')
    plt.ylabel('CO2 Emissions (billion metric tons)')
    plt.title(f'Polynomial Regression for CO2 Emissions in {state}')
    plt.legend()
    plt.grid(True)
    plt.show()
    print(f"\nPredicted CO2 Emissions for {state} (2024-2033):")
    for year, emission in zip(future_years.flatten(), predicted_emissions):
        print(f"{emission:.2f}")
```
